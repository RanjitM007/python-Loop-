# The while Loop
 - With the while loop we can execute a set of statements as long as a condition is true.
 

                            - image source GeekforGeeks

```python
#ex-1
a=1
while a <= 10:
    print(a)
    a+=1
    #a=a+1
```

```python
#ex2
b=1
while b < 20 :
    print("b is less than 20 →",b)
    b+=1
```

# The break Statement
 - With the break statement we can stop the loop even if the while condition is true

```python
#ex-1
k=1
while k < 15:
    print(k)
    if k==4:
        print("k value is → ",k)
        break
    k+=1
```

```python
#ex-1
k=1
while k < 15:
    print(k)
    if k==4:
        print("k value is → ",k)
        break
    k+=1
```

# The continue Statement
  - With the continue statement we can stop the current iteration, and continue with the next:

```python
#ex-1
k=1
while k < 15:
    k+=1
    if k==4:
        continue 
    print(k)
   
```

# The else Statement

 - With the else statement we can run a block of code once when the condition no longer is true

```python
a=2

while a < 10 :
    print("global",a)
    print("Before increment ",a)
    a+=1
    print(a)
else:
        print(" a is gretter than 10 ")
     
```
